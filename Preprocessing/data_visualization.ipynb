4.Perform following function. use tips.csvdataset.

```python
tips_df = pd.read_csv("tips.csv")
```

```python
# a)Data Exploration for tips dataset
print(tips_df.head())

# Show basic statistics
print(tips_df.describe())

# Check for missing values
print(tips_df.isnull().sum())
```

```python
# b)Show Histogram for distribution of the “total_bill” variable
plt.figure(figsize=(10, 6))
# sns.histplot(data=tips_df, x='total_bill', kde=True)
plt.hist(tips_df['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.grid(True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# c)visualizing the relationship between two variables:“sex” and “total_bill” to see if there is a difference in the amount spent by men and women when dining out.

plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='total_bill', data=tips_df)
plt.title('Relationship Between Sex and Total Bill')
plt.xlabel('Sex')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
# d)Create a box plotin seabornto further understand the difference between the amount spent by male and female customers.Interpret the result.
g = sns.boxplot(x='sex', y='total_bill', data=tips_df)
g.set_title('Distribution of Total Bill by Sex')
g.set_xlabel('Sex')
g.set_ylabel('Total Bill ($)')
plt.show()
```

```python
# e)Creating a pair plotin Seaborn to visualize the relationship between every numeric variable in the dataset at once
plt.figure(figsize=(10, 8))
sns.set(style="ticks")
sns.pairplot(tips_df)
plt.show()
```

5.Load any dataset of your choiceor create your own data. Perform the following

a)Create multiple line chartwith required thickness. Make subplot&Plot subplot

b)Set title to subplots.

c)Draw a horizontal bar chart.

d)Create a Scatter Plot with several colors

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 15))
flights = sns.load_dataset("flights")


# Plot flights dataset
sns.lineplot(ax=axes[0], x="year", y="passengers", hue="month", data=flights)
axes[0].set_title('Flights Dataset - Yearly Passengers')


# Plot passengers by month and year
sns.lineplot(ax=axes[1], x="year", y="passengers", hue="month", data=flights)
axes[ 1].set_title('Passengers by Month and Year')
axes[ 1].legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="month", y="passengers", data=flights[::10], orient='h')
plt.title('Monthly Passenger Count')
plt.xlabel('Number of Passengers')
plt.ylabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=flights, x="passengers", y="year", hue="month", palette=sns.color_palette(n_colors=len(set(flights['month']))))
plt.title('Scatter Plot of Passengers vs Year by Month')
plt.xlabel('Number of Passengers')
plt.ylabel('Year')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

6.You are given a dataset of test scores from three different classes. Your task is to create a box plot to visualize the distribution of test scores across these classes. The goal is to identify any differences in score distributions, spot potential outliers, and understand the central tendency and variability of the scores in each class.Explain the output plot.

```python
data = {'Class': ['A'] * 10 + ['B'] * 10 + ['C'] * 10,
        'Score': [82, 85, 88, 90, 91, 87, 84, 92, 86, 83, 78, 80, 77, 79, 81, 75, 76, 82, 81, 78, 90, 92, 94, 91, 89, 95, 93, 92, 97, 98]}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Score', data=df)


plt.title('Distribution of Test Scores Across Classes')
plt.xlabel('Class')
plt.ylabel('Test Score')
plt.show()
```


    Overall Distribution: The box plot shows the distribution of test scores for each class (A, B, and C).

    Central Tendency:
        Class A: The box represents the middle 50% of the data .
        Class B: The box is slightly shorter, indicating less spread in the data.
        Class C: The box is longer, suggesting more variability in the scores.

    Outliers:
        Class A has one outlier (score > 90).
        Class B has no visible outliers.
        Class C has two outliers (scores > 90).

    Median:
        Class A: Approximately 85-90
        Class B: Approximately 80-85
        Class C: Approximately 90-95

    Interquartile Range (IQR):
        Class A: About 5-10
        Class B: About 5-10
        Class C: About 10-15

    Variability:
        Class A shows moderate variability.
        Class B shows relatively low variability.
        Class C shows the highest variability among the three classes.
