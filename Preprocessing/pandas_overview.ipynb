4) Write a Pandas program to create a line plot of the historical stock prices
of Alphabet Inc. between two specific dates. Extract data from
alphabet_stock_data.csv file.

```python
alpha_df = pd.read_csv("alphabet_stock_data.csv")
alpha_df
```

```python
import matplotlib.pyplot as plt

st = '2020-04-01'
end = '2020-04-04'


filter_df = alpha_df[(alpha_df['Date'] >= st) & (alpha_df['Date'] <= end)]
# filtering dataset for a particular range of dates and plotting it using matplotlib
plt.plot(filter_df['Date'], filter_df['Close'])
plt.title('Alphabet Inc. Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

6) Sales Data Analysis
Scenario: You have a dataset containing sales transactions for a retail store. The
dataset includes columns for the transaction date, product category, sales amount,
and store location. Your task is to calculate the total sales amount for each product
category and identify the top 3 product categories with the highest sales

```python
data = {'transaction_date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
'product_category': ['Electronics', 'Furniture', 'Electronics', 'Clothing', 'Furniture'],
'sales_amount': [500.00, 150.00, 200.00, 100.00, 200.00],
'store_location': ['New York', 'Chicago', 'New York', 'Chicago', 'New York']}

trans_df = pd.DataFrame(data)
trans_df
```

```python
total_sales = trans_df.groupby('product_category')['sales_amount'].sum()
total_sales = total_sales.sort_values(ascending=False)
# sorting in reverse to get the highest sales amount and printing the top 3
total_sales.head(3)
```

7) Customer Churn Analysis
Scenario: You have a dataset of customer subscriptions where each row represents
a subscription with columns for customer ID, subscription start date, subscription
end date, and subscription status. Your task is to calculate the churn rate, which
is the percentage of customers who have canceled their subscriptions.

```python
data = {'customer_id': [101, 102, 103, 104, 105],
'subscription_start_date': ['2023-01-01', '2023-06-15', '2023-03-22', '2022-12-01', '2023-05-10'],
'subscription_end_date': ['2024-01-01', '2024-06-15', '2023-09-22', '2023-12-01', None],
'status': ['Active', 'Active', 'Canceled', 'Canceled', 'Active']}

cust_df = pd.DataFrame(data)
cust_df
```

```python
total_customers = cust_df['customer_id'].count()
churned_customers = cust_df[cust_df['status'] == 'Canceled']['customer_id'].count()

churn_rate = (churned_customers / total_customers) * 100
# calculating the churn rate
churn_rate
```
