# Upper Confidence Bound (UCB)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing UCB

```
import math
N = 10
d = 10
ads_selected = []
numbers_of_selections = [0] * d
sums_of_rewards = [0] * d
total_reward = 0

for n in range(0,N):
  ad = 0
  max_upper_bound = 0
  print("Ad is ",ad)
  print("Max Upper Bound is ",max_upper_bound)
  for i in range(0,d):
    if numbers_of_selections[i] > 0:
      average_reward = sums_of_rewards[i] / numbers_of_selections[i]
      delta_i = math.sqrt( 3/2 * math.log(n+1) / numbers_of_selections[i] )
      upper_bound = average_reward + delta_i
      print("Average Reward is ",average_reward)
      print("Delta is ",delta_i)
      print("Upper Bound is ",upper_bound)
      print("Sum of Reward is ",sums_of_rewards[i],"No of selections are ",numbers_of_selections[i])
      print("No of selections are",numbers_of_selections)
    else:
      upper_bound = 1e400
    if(upper_bound > max_upper_bound):
      max_upper_bound = upper_bound
      ad = i
  ads_selected.append(ad)
  numbers_of_selections[ad] +=1
  reward = dataset.values[n,ad ]
  sums_of_rewards[ad] +=reward
  total_reward += reward
  print("Ads Selected " ,ads_selected)
  print("Reward ",reward)
  print("Sum of Rewards ",sums_of_reward)
  print("Total Reward",total_reward)
```

## Visualising the results

```
plt.hist(ads_selected)
plt.title('Histogram of Ads Selection')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected ')
plt.show()
```
